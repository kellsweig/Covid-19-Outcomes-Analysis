### EDA merged state with health

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
```

```python
# read in Data
covid_df = pd.read_csv('Data/merged_state_with_health.csv')
covid_df.head()
```

```python
covid_df.columns
```

```python
covid_df.isna().sum()
```

```python
covid_df['avg_deaths_2020-2022'] = covid_df[['Exc_deaths_2020','Exc_deaths_2021','Exc_deaths_2022']].mean(axis=1)
```

```python
covid_df['Mask_Mandate']
```

```python
covid_df['Mask_Mandate'].fillna('12/31/2020',inplace=True)
```

```python
summary_stats = covid_df.describe()
```

```python
summary_stats
```

```python
covid_df.head()
```

```python
####create histogram for all numerical columns

num_columns = covid_df.select_dtypes(include=['int64','float64'])
                                              
for column in num_columns.columns:
    plt.figure(figsize=(5,5))
    plt.hist(num_columns[column], bins=40,color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('frequency')
    
```

```python
#### making a correlation map

correlation_matrix = covid_df.corr()
avg_deaths_correlation = correlation_matrix['avg_deaths_2020-2022']
plt.figure(figsize=(30, 25))
sns.heatmap(avg_deaths_correlation.to_frame(), annot=True, cmap='coolwarm', cbar=True, linewidths=0.5)
plt.title("Correlation Heatmap for avg_deaths_2020-2022")
plt.show()
```

### EDA on Merged state final

```python
#read data
state_df = pd.read_csv('Data/merged_state_final.csv')
```

```python
state_df
```

```python
covid_by_pop = state_df['covid_2020'] + state_df['covid_2021'] + state_df['covid_2022']
state_df['covid_deaths_by_population'] = (covid_by_pop / state_df['2020 Population']) * 10000

```

```python
state_df.head()
```

```python
bottom_10_states = state_df.sort_values(by='covid_deaths_by_population', ascending=True).head(10)
```

```python
bottom_10_states
```

```python
top_10_states = state_df.sort_values(by='covid_deaths_by_population', ascending=False).head(10)
top_10_states
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(state_df['vax_percent'], state_df.index, alpha=0.6)
plt.title('Scatter Plot of Vaccination Percent by State')
plt.xlabel('Vaccination Percent')
plt.ylabel('State Index')
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot for 2020
plt.scatter(state_df['Inc_Per_Cap_2020'], state_df.index, label='2020', alpha=0.6)

# Scatter plot for 2021
plt.scatter(state_df['Inc_Per_Cap_2021'], state_df.index, label='2021', alpha=0.6)

# Scatter plot for 2022
plt.scatter(state_df['Inc_Per_CAp_2022'], state_df.index, label='2022', alpha=0.6)

plt.title('Scatter Plot of Income Per Capita by State (2020-2022)')
plt.xlabel('Income Per Capita')
plt.ylabel('State Index')
plt.legend()
```
